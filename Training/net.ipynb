```python
# univariate cnn example
from numpy import array
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Conv1D
from tensorflow.keras.layers import MaxPooling1D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import load_model
import pandas as pd
import numpy as np
import keras
```

```python
c = pd.read_csv("resNew.csv")

arr = np.empty((12393,81),dtype='float32')

print(arr.shape)

i=0
n=0

for riga in c["state"]:
    n=0    
    for el in riga:
        arr[i,n]=el        
        n+=1
    i+=1    
    
arr= arr/3
print(arr)
```

```python
assert not np.any(np.isnan(arr))
```

```python
out = np.empty((12393),dtype=object)

i=0

max=-100
min=100

for i in range(12393):
    out[i]=(c['w'][i] - c['b'][i])
    if(out[i]>max):
        print(out[i])
        max=out[i]
    if(out[i]<min):
        min=out[i]
        
interval=max-min

print(min)
print(interval)
        

##meanValue=np.mean(out)

##std=np.std(out)



##print(str(meanValue))
##print(std)

for i in range(12393):
    out[i]=2*((out[i]-min)/interval)-1
    

out=out.astype('float32')

print(out)

```

```python
assert not np.any(np.isnan(out))
```

```python
# choose a number of time steps
n_steps = 3
# reshape from [samples, timesteps] into [samples, timesteps, features]
n_features = 1
X = arr.reshape((arr.shape[0], arr.shape[1], n_features))
# define model
model = Sequential()
model.add(Conv1D(filters=64, kernel_size=2, activation='relu', input_shape=(81, n_features)))
model.add(MaxPooling1D(pool_size=2))
model.add(Flatten())
model.add(Dense(50, activation='relu'))
model.add(Dense(1))
opt = Adam(learning_rate=0.0001)
model.compile(optimizer=opt, loss='mse')
# fit model
mcp_save = keras.callbacks.ModelCheckpoint('net.h5', save_best_only=True, monitor='val_loss')
model.fit(X, out, epochs=250, verbose=1, validation_split = 0.3, callbacks=[mcp_save])
model.save('net.h5')

```
